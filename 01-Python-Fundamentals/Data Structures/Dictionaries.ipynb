# Introduction to Dictionaries

A **dictionary** is a built-in Python data structure that stores data in **key-value pairs**. Each key acts as a unique identifier that maps to a corresponding value, allowing data to be retrieved quickly and efficiently.

Unlike lists and tuples, dictionaries do not access elements by their position (index). Instead, they use **keys** to access values.

### Characteristics of Dictionaries

- Store data as **key-value pairs**.
- Keys must be **unique**.
- Keys must be **immutable** (such as strings, numbers, or tuples containing immutable elements).
- Values can be of **any data type**.
- Dictionaries are **mutable**, meaning you can add, update, or remove key-value pairs after creation.
- Dictionaries preserve insertion order in Python 3.7 and later.
- Dictionary lookups, insertions, and deletions have an average time complexity of **O(1)** due to hashing.

### Example

```python
student = {
    "name": "Siva",
    "age": 23,
    "city": "Kurnool"
}

print(student["name"])
```

**Output**

```
Siva
```

````python
student = {"name":"siva","age":22,"grade":24}
print(student)
print(type(student))
````

````python
## Single key is always used

student={"name":"siva","age":22,"name":23}
print(student)
````

````python
## Accessing Dictionary Elements
student = {"name":"siva","age":22,"grade":"A"}
print(student)
````

````python
## Accessing Dictionary Elements

print(student['grade'])
print(student['age'])

## Accessing using get() Methods

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
````

````python
## Modifying Dictionary Elements
## Dictionary are mutable, so you can add, update or delete elements
print(student)
````

````python
student["age"] = 24  ## Update value for the key
print(student)

student["address"] = "India" ## Added a new key and value
print(student)
````

````python
del student['grade'] ## delete key and value pair
print(student)
````

````python
## Dictionary Methods

keys = student.keys()
print(keys)
values = student.values() ## get all values
print(values)

items = student.items() ## get all key value pairs
print(items)
````

````python
## Shallow copy

student_copy = student
print(student)
print(student_copy)
````

````python
student["name"] ="siva2"
print(student)
print(student_copy)
````

````python
student_copy1=student.copy()  ## shallow copy
print(student_copy1)
print(student)
````

````python
student["name"] = "Siva3"
print(student_copy1)
print(student)
````

````python
## Iterating Over Dictionaries
## You can use loops to iterate over dictionaries , keys, values, or items

## Iterating over keys
for keys in student.keys():
    print(keys)
````

````python
## iterate over values
for value in student.values():
    print(value)
````

````python
## iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
    
````

````python
## Nested Dictionaries

student={
    "student1":{"name":"siva","age":22},
    "student2":{"name":"teju","age":21},

}
print(student)
````

````python
## Accessing nested dictionaries elements
print(student["student2"]["name"])
print(student["student2"]["age"])
````

````python
## Iterating over nested dictionaries
for student_id, student_info in student.items():
    print(f"{student_id}:{student_info}")
    for key, value in student_info.items():
        print(f"{keys}:{value}")
        
````

````python
## Dictionary Comprehension
squares = {x:x**2 for x in range(5)}
print(squares)
````

````python
## Condition Dictionary comprehension
evens = {x:x**2 for x in range(10) if x%2==0}
print(evens)
````

````python
## Practical Examples

## Use a Dictionary to count the frequency of elements in list

numbers=[1,2,3,4,4,4,5,5,6,6,7,7]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number] = 1
print(frequency)
````

````python
## Merge 2 dictionaries into one

dict1 = {"a":1,"b":2}
dict2 = {"b":3,"c":4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
````
