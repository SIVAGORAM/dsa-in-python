Sets

Sets are a built in data type in python used to store collection of unique items. they are unordered , meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

## 22. Key Takeaways

- Sets store unique elements.
- Sets are unordered and mutable.
- Duplicate values are automatically removed.
- Sets provide O(1) average lookup time.
- Set operations are powerful and efficient.
- Sets are fundamental for hashing-based DSA problems.

````python
## create a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
````

````python
my_empty_set=set()
print(type(my_empty_set))
````

````python
my_empty_set=set([1,2,3,4,5,6,7])
print(type(my_empty_set))
print(my_empty_set)
````

````python
## Set not allow duplicate elements
my_empty_set=set([1,2,3,4,5,6,6,6,7])

print(my_empty_set)
````

````python
## Basics sets operations
## Adding and removing elements

my_set.add(100)
print(my_set)
````

````python
## Remove the element from a set
my_set.remove(3)
print(my_set)
````

````python
my_set.discard(100)
print(my_set)
````

````python
## Pop Method
removed_element = my_set.pop()
print(removed_element)
print(my_set)
````

````python
## Clear all the elements
my_set.clear()
print(my_set)
````

````python
## Set Membership Test
my_set={1,2,3,4,4,5}
print(3 in my_set)
print(10 in my_set)
````

````python
## Mathematical Operations

set1={1,2,3,4,5,6}
set2={4,5,5,6,7,7,88,9,}

## union 
union_set = set1.union(set2)
print(union_set)

## Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)
````

````python
set1={1,2,3,4,5,6}
set2={4,5,5,6,7,7,88,9,}

## Difference
print(set1.difference(set2))
````

````python
set1
````

````python
set2.difference(set1)
````

````python
## Symmetric Difference

set1.symmetric_difference(set2)
````

````python
## Sets Methods
set1 = {1,2,3,4}
set2= {3,4,5,6,7}

## is subset
print(set1.issubset(set2))

print(set1.issuperset(set2))
````

````python
lst =[1,2,3,4,5,5,6,6,6]
set(lst)
````
