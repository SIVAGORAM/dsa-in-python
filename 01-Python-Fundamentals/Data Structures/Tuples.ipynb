Introduction to tuples

- Tuples are ordered collection of items that are immutable. they are similar to lists, but their immutability makes them different.


````python
## Creating a Tuples
empty_tuples = ()
print(empty_tuples)
print(type(empty_tuples))
````

````python
lst = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
````

````python
numbers = tuple([1,2,3,4,5,6])
numbers
````

````python
list((1,2,3,4,5,6))
````

````python
mixed_tuple = (1,"Hello world",3.14, True)
print(mixed_tuple)
````

````python
## Accessing tuples Elements
numbers
````

````python
print(numbers[2])
````

````python
print(numbers[0:4])
````

````python
print(numbers[::-1])
````

````python
## Tuple Operations

concatenation_tuple = numbers + mixed_tuple
print(concatenation_tuple)
````

````python
mixed_tuple * 3
````

````python
numbers * 3
````

````python
## Immutable natures of tuples
## tuples are immutable, meaning their elements cannot be changed once assigned

lst = [1,2,3,4,5]
print(lst)

lst[1]="siva"
print(lst)
````

````python
## Tuple Method:-

print(numbers.count(1))
print(numbers.index(3))
````

````python
## Packing and Unpacking Tuples

packed_tuple = 1,"Hello",3.14
print(packed_tuple)
````

````python
## Unpacking a tuple

a,b,c=packed_tuple
print(a)
print(b)
````

````python
## Unpacking with *
numbers = (1,2,3,4,5,6)
first,*middle, last = numbers
print(first)
print(middle)
print(last)
````

````python
## Nested tuple
lst = [[1,2,3,4],[6,7,8,9],"Hello",3.14,"c"]
lst[0][0:3]
````

````python
lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
````

````python
nested_tuple = ((1,2,3),("a","b","c"),(True,False))

## Access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
````

````python
## Iterating over nester tuples

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end="")
    print()
````

Conclusion

- Tuples are versatile and useful in many real-world scenarios where an immutable and ordered  collection of items is required. they are commonly used in data structured, function arguments and return values, and as dictionary keys. understanding how to leverage tuples effectively can improve the efficiency and readability of your python code.
