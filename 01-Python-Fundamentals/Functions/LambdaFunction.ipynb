# Lambda Functions in Python

## Definition

A **lambda function** is a small, anonymous function in Python that is defined using the `lambda` keyword. Unlike regular functions created with `def`, a lambda function does not have a name unless it is assigned to a variable.

A lambda function can accept **any number of arguments**, but it can contain **only one expression**. The value of that expression is automatically returned.

Lambda functions are commonly used for short, simple operations and are often passed as arguments to higher-order functions such as `map()`, `filter()`, `reduce()`, and `sorted()`.

---

## Syntax

```python
lambda parameters: expression
```

---

## Example 1: Add Two Numbers

```python
add = lambda a, b: a + b

print(add(10, 20))
```

### Output

```
30
```

---

## Example 2: Square of a Number

```python
square = lambda x: x ** 2

print(square(5))
```

### Output

```
25
```

---

## Example 3: Find the Maximum Number

```python
maximum = lambda a, b: a if a > b else b

print(maximum(10, 20))
```

### Output

```
20
```

---

## Characteristics of Lambda Functions

- Anonymous (no function name is required).
- Defined using the `lambda` keyword.
- Can accept any number of arguments.
- Can contain only one expression.
- Automatically returns the result of the expression.
- Best suited for short, simple functions.
- Commonly used with higher-order functions.

---

## When to Use Lambda Functions

Use lambda functions when:

- The function is small and simple.
- The function is used only once.
- Passing a function as an argument.
- Working with `map()`, `filter()`, `reduce()`, or `sorted()`.

Avoid lambda functions for complex logic. In such cases, use a regular function with the `def` keyword.

---

## Lambda Function vs Regular Function

| Regular Function | Lambda Function |
|------------------|-----------------|
| Defined using `def` | Defined using `lambda` |
| Has a function name | Anonymous by default |
| Can contain multiple statements | Only one expression |
| Requires an explicit `return` (if returning a value) | Returns the expression automatically |
| Better for complex logic | Better for short operations |

---

## Key Takeaways

- A lambda function is a small anonymous function.
- It is created using the `lambda` keyword.
- It can accept multiple arguments but only one expression.
- The expression's result is returned automatically.
- It is commonly used with higher-order functions like `map()`, `filter()`, `reduce()`, and `sorted()`.
