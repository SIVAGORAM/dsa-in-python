# The `map()` Function in Python

## Definition

The **`map()`** function in Python applies a specified function to every element of an iterable (such as a list, tuple, or set) and returns a **map object**, which is an iterator containing the transformed values.

The `map()` function is commonly used when you want to perform the same operation on every element of an iterable without writing an explicit loop.

---

## Syntax

```python
map(function, iterable)
```

### Parameters

- **function**: The function to apply to each element of the iterable.
- **iterable**: The iterable (such as a list, tuple, set, or string) whose elements will be processed.

### Returns

- A **map object** (iterator) containing the transformed elements.
- You can convert the result into a list, tuple, or set using `list()`, `tuple()`, or `set()`.

---

## Example 1: Using a Normal Function

```python
def square(num):
    return num ** 2

numbers = [1, 2, 3, 4, 5]

result = map(square, numbers)

print(list(result))
```

### Output

```text
[1, 4, 9, 16, 25]
```

---

## Example 2: Using a Lambda Function

```python
numbers = [1, 2, 3, 4, 5]

result = map(lambda x: x ** 2, numbers)

print(list(result))
```

### Output

```text
[1, 4, 9, 16, 25]
```

---

## Example 3: Convert Strings to Uppercase

```python
names = ["siva", "ram", "krishna"]

result = map(str.upper, names)

print(list(result))
```

### Output

```text
['SIVA', 'RAM', 'KRISHNA']
```

---

## Example 4: Convert Celsius to Fahrenheit

```python
temperatures = [0, 25, 37, 100]

fahrenheit = map(lambda c: (c * 9/5) + 32, temperatures)

print(list(fahrenheit))
```

### Output

```text
[32.0, 77.0, 98.6, 212.0]
```

# Characteristics of `map()`

- Applies the same function to every element.
- Returns a **map object** (iterator).
- Evaluates elements lazily (only when needed).
- Works with any iterable.
- Often used with lambda functions.
- Can accept multiple iterables if the function expects multiple arguments.

---

## Advantages of `map()`

- Reduces the need for explicit loops.
- Produces cleaner and more readable code.
- Supports functional programming.
- Efficient because it returns an iterator instead of creating a new list immediately.

---

## When to Use `map()`

Use `map()` when:

- Every element needs the same transformation.
- You want concise and readable code.
- You're applying a function to an entire iterable.
- Working with functional programming techniques.

---

## When Not to Use `map()`

Avoid using `map()` when:

- The transformation requires complex logic.
- Readability suffers.
- A simple `for` loop is easier to understand.
- A list comprehension is more readable.

---

## `map()` vs List Comprehension

| `map()` | List Comprehension |
|---------|--------------------|
| Uses a function | Uses an expression |
| Returns a map object | Returns a list |
| Good for existing functions | Better for custom logic |
| Often used with lambda | More Pythonic for simple transformations |

## Key Takeaways

- `map()` applies a function to every element of an iterable.
- It returns a **map object** (an iterator).
- It works with regular functions and lambda functions.
- It is useful for transforming data efficiently.
- It is widely used in functional programming.

````python
def square(x):
    return x*x
square(10)
````

````python
numbers=[1,2,3,4,5,6,7,8]

map(square,numbers)
````

````python
numbers=[1,2,3,4,5,6,7,8]

list[map(square,numbers)]
````

````python
numbers=[1,2,3,4,5,6,7,8]

list(map(square,numbers))
````

````python
## Lambda Function with map
numbers = [1,2,3,4,5,6,7,8]
list(map(lambda x:x*x,numbers))
````

````python
## Map Multiple iterables

numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y,numbers1,numbers2))

print(added_numbers)
````

````python
## Map() to convert a list of strings to integers
# Use map to convert strings to integers

str_numbers = ['1','2','3','4','5']
int_numbers = list(map(int,str_numbers))

print(int_numbers) 
````

````python
words =['apple','banana','cherry']
upper_words=list(map(str.upper,words))
print(upper_words)
````

````python
def get_name(person):
    return person['name']

people=[
    {'name':'siva','age':23},
    {'name':'teju','age':22}
]
list(map(get_name,people))
````
