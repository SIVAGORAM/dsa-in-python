# Python Standard Library Overview

## Introduction

The **Python Standard Library** is a comprehensive collection of built-in modules and packages that comes bundled with Python. These modules provide a wide range of functionality, allowing you to perform common programming tasks without installing additional software.

The Standard Library includes modules for working with:

- Mathematics
- Random number generation
- Dates and time
- File and directory operations
- Operating system interactions
- Data structures
- String manipulation
- Regular expressions
- JSON and CSV processing
- Networking
- Multithreading
- And much more

Because these modules are included with Python, you can use them immediately by importing them into your program.

---

## Why Use the Standard Library?

The Python Standard Library helps developers:

- Reuse well-tested and reliable code.
- Reduce development time.
- Avoid writing common functionality from scratch.
- Improve code readability and maintainability.
- Build powerful applications with minimal effort.

---

## Some Commonly Used Standard Library Modules

| Module | Purpose |
|---------|---------|
| `math` | Mathematical functions and constants |
| `random` | Generate random numbers and make random selections |
| `datetime` | Work with dates and times |
| `time` | Measure and manipulate time |
| `os` | Interact with the operating system |
| `sys` | Access Python interpreter and system-specific information |
| `statistics` | Perform statistical calculations |
| `collections` | Specialized container data types |
| `itertools` | Efficient tools for working with iterators |
| `functools` | Higher-order functions and utilities |
| `re` | Regular expressions for pattern matching |
| `json` | Read and write JSON data |
| `csv` | Read and write CSV files |
| `pathlib` | Object-oriented file system paths |
| `shutil` | High-level file operations |

````python
import array

arr = array.array('i', [1, 2, 3, 4, 5])

print(arr)
````

````python
import math
print(math.sqrt(16))
print(math.pi)
````

````python
## Random

import random
print(random.randint(1,10))
print(random.choice(['apple','banana','cherry']))
````

````python
## OS:- FIle and Directory Access
import os
print(os.getcwd())
````

````python
os.mkdir('test_dir')
````

````python
## High level Operations on filters and collection of files

import shutil
shutil.copyfile('source.txt','destination.txt')
````

````python
## Date Serialization

import json
data ={'name':'siva','age':22}

json_str=json.dumps(data)
print(json_str)
print(type(json_str))

parsed_data = json.loads(json_str)
print(parsed_data)
print(type(parsed_data))
````

````python
## CSV

import csv

with open('example.csv',mode='w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['siva',22])

with open('example.csv',mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
````

````python
## datetime
from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
print(yesterday)
````

````python
## time
import time

print(time.time())
time.sleep(2)
print(time.time())
````

````python
## Regular expressions

import re

pattern=r'\d+'
text='there are 123 apples 456'
match=re.search(pattern,text)
print(match.group())
````
