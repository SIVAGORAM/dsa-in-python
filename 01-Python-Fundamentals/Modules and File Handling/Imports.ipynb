# Importing Modules and Packages in Python

## Introduction

Python provides **modules** and **packages** to help organize, manage, and reuse code efficiently.

Instead of writing all your code in a single file, you can divide it into multiple modules and packages, making your programs more organized, readable, maintainable, and reusable.

Python offers many built-in modules, and you can also create your own modules or install third-party packages using **pip**.

In this chapter, you'll learn how to:

- Import built-in modules
- Create your own modules
- Import specific functions or variables
- Use aliases with `as`
- Import everything using `*`
- Understand packages
- Install and use third-party libraries
- Follow best practices for importing modules

---

# What is a Module?

A **module** is a single Python file (`.py`) that contains reusable code such as:

- Variables
- Functions
- Classes
- Constants

Modules help organize code into smaller, manageable files that can be imported and reused in other Python programs.

### Examples of Built-in Modules

- `math`
- `random`
- `os`
- `sys`
- `datetime`
- `statistics`
- `collections`

# What is a Package?

A **package** is a collection of related Python modules organized inside a directory.

Packages help organize large projects by grouping similar modules together.

A package usually contains an `__init__.py` file (optional in modern Python versions) to indicate that the directory is a Python package.

Example:

```
my_package/
│
├── __init__.py
├── math_utils.py
├── string_utils.py
└── file_utils.py
```

## Key Takeaways

- A **module** is a single Python file (`.py`) containing reusable code.
- A **package** is a directory that contains multiple related modules.
- Modules improve code reusability and organization.
- Packages help structure large applications.
- Python includes many built-in modules, and thousands of third-party packages are available through **pip**.

````python
import math
math.sqrt(16)
````

````python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
````

````python
import numpy as np
np.array([1,2,3,4,5])
````

````python
from math import *
print(sqrt(16))
print(pi)
````

````python
## Importing package from another package folder from package/maths.py importing math function here.
from package.maths import addition
addition(2,3)
````

````python
## Another way to import packages
from package import maths
maths.addition(2,3)
````
