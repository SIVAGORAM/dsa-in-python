# Magic Methods (Dunder Methods) in Python

## Introduction

**Magic Methods**, also known as **Dunder Methods (Double Underscore Methods)**, are special predefined methods in Python that **start and end with double underscores (`__`)**.

These methods allow you to define or customize the behavior of objects for built-in operations such as:

- Object creation
- Printing objects
- Arithmetic operations
- Comparison operations
- Iteration
- Length calculation
- String representation
- Operator overloading
- And many more...

The word **"Dunder"** simply means **Double Under** (Double Underscore).

For example:

```python
__init__
__str__
__repr__
__len__
__add__
__eq__
```

---

# Why Do We Need Magic Methods?

Normally, Python knows how to perform operations on built-in data types.

Example:

```python
print(10 + 20)
```

Output

```text
30
```

Python knows how to add integers.

Similarly,

```python
print("Hello " + "World")
```

Output

```text
Hello World
```

Python knows how to concatenate strings.

But what if we create our own class?

```python
class Student:

    def __init__(self, name):
        self.name = name


student1 = Student("Siva")
student2 = Student("Teju")

print(student1 + student2)
```

Output

```text
TypeError:
unsupported operand type(s) for +: 'Student' and 'Student'
```

Python doesn't know **how to add two Student objects**.

Magic methods solve this problem.

---

# Definition

**Magic Methods** are special predefined methods that Python automatically calls when performing built-in operations on objects.

They allow developers to customize the behavior of their classes.

| Magic Method | Description |
|--------------|-------------|

| `__str__()` | Returns a user-friendly string representation |
| `__repr__()` | Returns an official string representation |
| `__len__()` | Returns the length of an object |
| `__add__()` | Defines behavior for the `+` operator |
| `__sub__()` | Defines behavior for the `-` operator |
| `__mul__()` | Defines behavior for the `*` operator |
| `__truediv__()` | Defines behavior for the `/` operator |

| `__lt__()` | Defines behavior for the `<` operator |

# 1. `__init__()` Method

## Definition

The `__init__()` method is called **automatically** whenever a new object is created.

It is known as the **Constructor**.

It is mainly used to initialize an object's attributes.

### Example

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age


student = Student("Siva", 22)

print(student.name)
print(student.age)
```

### Output

```text
Siva
22
```

---

# 2. `__str__()` Method

## Definition

The `__str__()` method returns a **human-readable string representation** of an object.

Whenever you use:

```python
print(object)
```

Python automatically calls the `__str__()` method.

### Without `__str__()`

```python
class Student:

    def __init__(self, name):
        self.name = name


student = Student("Siva")

print(student)
```

### Output

```text
<__main__.Student object at 0x000001A9E5C7A910>
```

This output is not user-friendly.

---

### With `__str__()`

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __str__(self):
        return f"Student Name: {self.name}"


student = Student("Siva")

print(student)
```

### Output

```text
Student Name: Siva
```

Now the object is displayed in a readable format.

---

# 3. `__repr__()` Method

## Definition

The `__repr__()` method returns the **official string representation** of an object.

It is mainly used for:

- Debugging
- Logging
- Developers

Whenever you type an object in the Python interpreter or call `repr(object)`, Python uses the `__repr__()` method.

### Example

```python
class Student:

    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"Student('{self.name}')"


student = Student("Siva")

print(repr(student))
```

### Output

```text
Student('Siva')
```

---

# Difference Between `__str__()` and `__repr__()`

| `__str__()` | `__repr__()` |
|--------------|--------------|
| User-friendly representation | Official representation |
| Used by `print()` | Used by `repr()` |
| Easy to read | Useful for debugging |
| Intended for end users | Intended for developers |

# Real-Time Example

Imagine you're building an **Employee Management System**.

```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return f"{self.name} earns ₹{self.salary}"

    def __repr__(self):
        return f"Employee('{self.name}', {self.salary})"


employee = Employee("Siva", 50000)

print(employee)

print(repr(employee))
```

### Output

```text
Siva earns ₹50000
Employee('Siva', 50000)
```

Here:

- `__str__()` gives a readable output for users.
- `__repr__()` gives a detailed representation useful for developers and debugging.

- `__str__()` provides a user-friendly string representation.

# Real-Time Coding Example: Employee Management System (Magic Methods)

One of the best ways to understand **Magic Methods (Dunder Methods)** is through a real-world **Employee Management System**.

Imagine you're developing software for a company to manage employee information.

Each employee has:

- Employee ID
- Name
- Department
- Salary

Normally, if we print an employee object, Python displays the object's memory address, which is not useful.

Using **Magic Methods**, we can customize how the object behaves.

---

# Step 1: Create the Employee Class

```python
class Employee:

    def __init__(self, emp_id, name, department, salary):
        """
        __init__() is automatically called whenever
        an object is created.
        """

        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.salary = salary
```

---

# Step 2: Create an Object

```python
employee = Employee(
    101,
    "Siva",
    "Software Engineer",
    50000
)
```

---

# Step 3: Print the Object (Without __str__)

```python
print(employee)
```

### Output

```text
<__main__.Employee object at 0x000001F4A0F3A4F0>
```

This output is not useful.

It only tells us the object's memory location.

---

# Step 4: Add __str__()

```python
class Employee:

    def __init__(self, emp_id, name, department, salary):

        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.salary = salary

    def __str__(self):

        return (
            f"Employee ID : {self.emp_id}\n"
            f"Name        : {self.name}\n"
            f"Department  : {self.department}\n"
            f"Salary      : ₹{self.salary}"
        )
```

---

# Create Object

```python
employee = Employee(
    101,
    "Siva",
    "Software Engineer",
    50000
)
```

---

# Print Object

```python
print(employee)
```

### Output

```text
Employee ID : 101
Name        : Siva
Department  : Software Engineer
Salary      : ₹50000
```

Much better!

Instead of the memory address, Python now prints meaningful information.

---

# What Happens Internally?

When we write

```python
print(employee)
```

Python automatically calls

```python
employee.__str__()
```

Internally,

```
print(employee)

        │

        ▼

employee.__str__()

        │

        ▼

Returns String

        │

        ▼

Prints on Screen
```

You never call `__str__()` directly.

Python does it automatically.

---

# Step 5: Add __repr__()

Now let's make the object useful for developers.

```python
class Employee:

    def __init__(self, emp_id, name, department, salary):

        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.salary = salary

    def __str__(self):

        return (
            f"Employee ID : {self.emp_id}\n"
            f"Name        : {self.name}\n"
            f"Department  : {self.department}\n"
            f"Salary      : ₹{self.salary}"
        )

    def __repr__(self):

        return (
            f"Employee("
            f"{self.emp_id}, "
            f"'{self.name}', "
            f"'{self.department}', "
            f"{self.salary})"
        )
```

---

# Using repr()

```python
employee = Employee(
    101,
    "Siva",
    "Software Engineer",
    50000
)

print(repr(employee))
```

### Output

```text
Employee(101, 'Siva', 'Software Engineer', 50000)
```

---

```
repr(employee)

        │

        ▼

employee.__repr__()

        │

        ▼

Returns Official Representation

        │

        ▼

Displays Result
```

# Complete Example

```python
class Employee:

    def __init__(self, emp_id, name, department, salary):

        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.salary = salary

    def __str__(self):

        return (
            f"Employee ID : {self.emp_id}\n"
            f"Name        : {self.name}\n"
            f"Department  : {self.department}\n"
            f"Salary      : ₹{self.salary}"
        )

    def __repr__(self):

        return (
            f"Employee("
            f"{self.emp_id}, "
            f"'{self.name}', "
            f"'{self.department}', "
            f"{self.salary})"
        )


employee = Employee(
    101,
    "Siva",
    "Software Engineer",
    50000
)

print("Using print()")
print(employee)

print()

print("Using repr()")
print(repr(employee))
```

---

# Output
