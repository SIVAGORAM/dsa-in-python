# Abstraction

## Definition

**Abstraction** is one of the four fundamental principles of **Object-Oriented Programming (OOP)**.

Abstraction is the process of **hiding the complex implementation details** and exposing **only the essential features** or functionalities of an object to the user.

It allows users to interact with an object without knowing how the object performs its internal operations.

In simple words,

> **Abstraction means showing only what is necessary and hiding how it is implemented.**

---

## Why Do We Need Abstraction?

Without abstraction, users would need to understand every internal detail of a system before using it.

Abstraction helps us by:

- Hiding unnecessary implementation details.
- Reducing code complexity.
- Improving code readability.
- Increasing security by hiding internal logic.
- Making applications easier to use and maintain.
- Allowing developers to modify internal implementation without affecting users.

---

## How Abstraction Works in Python

Python implements abstraction using the **`abc` (Abstract Base Class)** module.

It provides:

- `ABC` (Abstract Base Class)
- `@abstractmethod`

These allow us to create abstract classes and abstract methods.

The child classes are responsible for providing the implementation of abstract methods.

---

## Example

```python
from abc import ABC, abstractmethod


class Animal(ABC):

    @abstractmethod
    def speak(self):
        pass


class Dog(Animal):

    def speak(self):
        print("Dog says Woof")


dog = Dog()

dog.speak()
```

### Output

```text
Dog says Woof
```

- Improves code readability.
- Makes applications more secure.

- Makes code easier to maintain.
- Improves scalability.

---

## Disadvantages of Abstraction

- Adds additional design complexity.
- Can increase the number of classes.
- Slight learning curve for beginners.
- May require more planning before implementation.

---

## Key Takeaways

- Abstraction hides the internal implementation of an object.
- Users only interact with the essential features.
- It focuses on **what an object does**, not **how it does it**.
- Python supports abstraction using the **`abc`** module.
- Abstract classes cannot be instantiated directly.
- Child classes must implement all abstract methods.
- Abstraction reduces complexity and improves maintainability.

# Difference Between Encapsulation and Abstraction

## Encapsulation

**Question:** "Who can access the data?"

Focuses on **protecting data**.

Example:

```python
class BankAccount:

    def __init__(self):
        self.__balance = 10000
```

The balance is hidden using private variables.

---

## Abstraction

**Question:** "What should the user see?"

Focuses on **hiding implementation details**.

Example:

```python
atm.withdraw()
```

The user only sees the `withdraw()` method.

The internal implementation remains hidden.

---

# Simple Analogy

Imagine driving a car.

**You know:**

```
Start Car

Accelerate

Brake

Steering
```

**You don't know:**

```
Engine

Fuel Pump

Gearbox

Cooling System

Sensors

Battery Management
```

The car hides all the complexity and exposes only the controls you need.

This is exactly what **Abstraction** means:

> **Show only the essential features and hide the complex implementation details.**

One-line trick to remember:

- Encapsulation = Hide the data (using access modifiers like private variables).

- Abstraction = Hide the implementation (show only what the user needs to use).

# Real-Time Coding Example: ATM System (Abstraction)

One of the best real-world examples to understand **Abstraction** is an **ATM Machine**.

As a customer, when you want to withdraw money, you simply perform the following steps:

1. Insert ATM Card
2. Enter PIN
3. Enter Withdrawal Amount
4. Collect Cash

You don't know what happens internally.

The ATM internally performs many operations such as:

- Verifying the ATM Card
- Verifying the PIN
- Checking Account Balance
- Connecting to the Bank Server
- Deducting the Amount
- Dispensing Cash
- Updating the Database
- Printing the Receipt

All these implementation details are **hidden** from the customer.

The customer only interacts with the **Withdraw** functionality.

This is called **Abstraction**.

---

# Step 1: Create an Abstract Class

```python
from abc import ABC, abstractmethod

class ATM(ABC):

    @abstractmethod
    def withdraw(self, amount):
        pass
```

## Explanation

Here,

`ATM` is an **Abstract Class**.

It only tells us that every ATM should have a **withdraw()** method.

It does **not** tell us **how** money is withdrawn.

The implementation will be provided by the child class.

---

# Step 2: Implement the ATM

```python
from abc import ABC, abstractmethod

class ATM(ABC):

    @abstractmethod
    def withdraw(self, amount):
        pass


class SBIATM(ATM):

    def __init__(self):
        self.balance = 10000

    def withdraw(self, amount):

        print("Checking ATM Card...")

        print("Verifying PIN...")

        print("Connecting to Bank Server...")

        print("Checking Account Balance...")

        if amount <= self.balance:

            self.balance -= amount

            print(f"₹{amount} Withdrawn Successfully")

            print(f"Remaining Balance : ₹{self.balance}")

            print("Updating Database...")

            print("Printing Receipt...")

        else:

            print("Insufficient Balance")
```

---

# Step 3: Create an Object

```python
atm = SBIATM()
```

---

# Step 4: Withdraw Money

```python
atm.withdraw(3000)
```

---
