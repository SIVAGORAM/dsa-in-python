# Polymorphism in Python

## Introduction

**Polymorphism** is one of the four fundamental pillars of **Object-Oriented Programming (OOP)**. The word **Polymorphism** comes from two Greek words:

- **Poly** → Many
- **Morph** → Forms

So, **Polymorphism** means **"many forms"**.

In Object-Oriented Programming, polymorphism allows objects of different classes to be treated as objects of a common superclass. It enables the same interface (method or function) to perform different behaviors depending on the object that invokes it.

In Python, polymorphism allows us to write flexible, reusable, and extensible code by using the same method name for different implementations.

---

# What is Polymorphism?

**Polymorphism** is the ability of an object, method, or function to take multiple forms.

This means that the same method name can behave differently depending on the object that calls it.

In Python, polymorphism is primarily achieved through:

- Method Overriding
- Method Overloading (Limited Support)
- Duck Typing
- Operator Overloading
- Built-in Polymorphic Functions

---

# Why Do We Need Polymorphism?

Polymorphism helps developers to:

- Write reusable code.
- Reduce code duplication.
- Improve code flexibility.
- Increase code maintainability.
- Easily extend applications without modifying existing code.
- Follow the Open/Closed Principle (Open for extension, Closed for modification).

# 2. Method Overriding

## Definition

**Method Overriding** is a feature of Object-Oriented Programming (OOP) in which a child class provides its own implementation of a method that already exists in its parent class.

The method in the child class must have the **same name**, **same parameters**, and **same signature** as the method in the parent class.

When an object of the child class calls the method, the child class version is executed instead of the parent class version.

---

## Rules of Method Overriding

- The child class must inherit from the parent class.
- The method name must be the same as the parent class method.
- The method parameters should be the same.
- The child class provides a different implementation of the method.
- Python decides which method to call at runtime.

---

## Parent vs Child Methods

| Parent Class | Child Class |
|--------------|-------------|
| Defines the original method | Provides a new implementation |
| Generic behavior | Specific behavior |
| Can be overridden | Overrides the parent method |

---

## Using `super()`

The `super()` function is used to call the parent class method from the child class.

### Example

```python
class Animal:
    def speak(self):
        print("Animal speaks")


class Dog(Animal):
    def speak(self):
        super().speak()
        print("Dog barks")


dog = Dog()
dog.speak()
```

### Output

```text
Animal speaks
Dog barks
```

---

## Example

```python
class Animal:

        print("Animal speaks")
```

# 3. Method Overloading

## Definition

**Method Overloading** is the ability to define multiple methods with the same name but different parameters.

Unlike Java or C++, **Python does not support true method overloading** because it is a dynamically typed language.

Instead, Python achieves similar behavior using:

- Default arguments
- `*args`
- `**kwargs`

---

## Using Default Arguments

```python
class Math:

    def add(self, a, b=0, c=0):
        return a + b + c


obj = Math()

print(obj.add(10))
print(obj.add(10, 20))
print(obj.add(10, 20, 30))
```

### Output

```text
10
30
60
```

---

## Using `*args`

```python
class Math:

    def add(self, *numbers):
        return sum(numbers)


obj = Math()

print(obj.add(10))
print(obj.add(10, 20))
print(obj.add(10, 20, 30, 40))
```

### Output

```text
10
30
100
```

---

## Using `**kwargs`

```python
class Student:

    def details(self, **kwargs):
        print(kwargs)


obj = Student()

obj.details(name="Siva", age=23, city="Kurnool")
```

### Output

```text
{'name': 'Siva', 'age': 23, 'city': 'Kurnool'}
```

---

## Note

Python achieves method overloading using **default arguments**, **`*args`**, and **`**kwargs`** instead of defining multiple methods with the same name.

# 4. Duck Typing

## Definition

**Duck Typing** is a concept in Python where the type of an object is less important than the methods it implements.

If an object behaves like the expected object, Python accepts it regardless of its actual class.

> **"If it walks like a duck and quacks like a duck, then it is a duck."**

---

## Example

```python
class Duck:
    def speak(self):
        print("Quack")


class Dog:
    def speak(self):
        print("Bark")


def make_sound(animal):
    animal.speak()


make_sound(Duck())
make_sound(Dog())
```

### Output

```text
Quack
Bark
```

---

## Explanation

Python never checks whether the object is actually a `Duck` or a `Dog`.

It only checks whether the object has a `speak()` method.

---

# 5. Operator Overloading

## Definition

**Operator Overloading** allows operators such as `+`, `-`, `*`, `<`, `==`, etc., to work differently depending on the objects.

This is achieved using **Magic (Dunder) Methods**.

---

## Example

```python
class Number:

    def __init__(self, value):
        self.value = value

    def __add__(self, other):
        return Number(self.value + other.value)

    def __str__(self):
        return str(self.value)


n1 = Number(10)
n2 = Number(20)

print(n1 + n2)
```

### Output

```text
30
```

---

## Common Magic Methods

| Method | Operator |
|---------|----------|
| `__add__()` | + |
| `__sub__()` | - |
| `__mul__()` | * |
| `__truediv__()` | / |
| `__eq__()` | == |
| `__lt__()` | < |
| `__gt__()` | > |
| `__str__()` | print() |

# 7. Runtime Polymorphism

## Definition

**Runtime Polymorphism** is achieved through **Method Overriding**.

Python determines which method to execute while the program is running.

---

## Example

```python
class Animal:
    def speak(self):
        print("Animal speaks")


class Dog(Animal):
    def speak(self):
        print("Dog barks")


class Cat(Animal):
    def speak(self):
        print("Cat meows")


class Cow(Animal):
    def speak(self):
        print("Cow moos")


animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.speak()
```

### Output

```text
Dog barks
Cat meows
Cow moos
```

# 8. Compile-Time Polymorphism

## Definition

Compile-Time Polymorphism means the method to execute is determined during compilation.

Python **does not support true compile-time polymorphism** because it is a **dynamically typed language**.

Languages like Java and C++ support compile-time polymorphism through method overloading.

In Python, similar behavior is achieved using:

- Default arguments
- `*args`
- `**kwargs`

---

# 9. Abstract Classes & Polymorphism

## Definition

An **Abstract Class** is a class that cannot be instantiated and is used as a blueprint for other classes.

Abstract classes are created using the `abc` module.

---

## Example

```python
from abc import ABC, abstractmethod


class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


class Circle(Shape):

    def area(self):
        return "Circle Area"


circle = Circle()

print(circle.area())
```

### Output

```text
Circle Area
```

---

# 10. Interface-like Polymorphism

Python does not support interfaces like Java.

Instead, Python uses **Abstract Base Classes (ABC)** to achieve similar behavior.

An abstract class defines methods that child classes must implement.

# How Does Polymorphism Work?

Notice that we are calling the **same methods** for every object.

```python
animal.speak()
animal.eat()
animal.move()
```

Python automatically decides **which implementation to execute** based on the object's actual class.

```
Dog object
    ↓
Dog.speak()

Fish object
    ↓
Fish.speak()

Bird object
    ↓
Bird.speak()
```

The same applies to `eat()` and `move()`.

This decision happens **at runtime**, which is why it is called **Runtime Polymorphism**.
