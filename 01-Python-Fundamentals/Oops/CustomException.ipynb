# Custom Exceptions in Python (Raise and Throw an Exception)

## Introduction

In Python, there are many built-in exceptions such as:

- `ValueError`
- `TypeError`
- `IndexError`
- `KeyError`
- `ZeroDivisionError`
- `FileNotFoundError`

However, in real-world applications, these built-in exceptions are not always enough.

Sometimes we need to create our **own exceptions** to handle business-specific rules.

These are called **Custom Exceptions**.

---

# What is a Custom Exception?

A **Custom Exception** is a user-defined exception that is created by inheriting from Python's built-in `Exception` class.

It allows us to define our own error messages based on business requirements.

---

# Why Do We Need Custom Exceptions?

Imagine you're developing a:

- Banking Application
- ATM Machine
- E-Commerce Website
- Hospital Management System
- Student Portal

These systems have their own business rules.

For example,

- Minimum withdrawal amount
- Maximum order quantity
- Minimum age requirement
- Password validation
- Attendance percentage

These errors are not built into Python.

So we create our own exceptions.

---

# Syntax

```python
class CustomException(Exception):
    pass
```

Now we can raise this exception whenever required.

---

# The `raise` Keyword

The **`raise`** keyword is used to **throw an exception manually**.

Syntax

```python
raise ExceptionName("Error Message")
```

Example

```python
raise ValueError("Invalid Age")
```

Output

```text
ValueError: Invalid Age
```

# Real-Time Example 1: Bank Account (Minimum Balance)

Suppose a bank has a rule:

> The minimum account balance should always be **₹1000**.

If the balance goes below ₹1000, the bank should raise an exception.

---

## Step 1: Create a Custom Exception

```python
class InsufficientBalanceError(Exception):
    pass
```

---

## Step 2: Create the Bank Account Class

```python
class BankAccount:

    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):

        if self.balance - amount < 1000:

            raise InsufficientBalanceError(
                "Minimum balance of ₹1000 must be maintained."
            )

        self.balance -= amount

        print(f"₹{amount} Withdrawn Successfully")
        print(f"Remaining Balance : ₹{self.balance}")
```

---

## Step 3: Create an Object

```python
account = BankAccount(5000)
```

---

## Step 4: Withdraw Money

```python
account.withdraw(4500)
```

---

## Output

```text
InsufficientBalanceError:
Minimum balance of ₹1000 must be maintained.
```

---

# What Happened?

The balance is

```
₹5000
```

Trying to withdraw

```
₹4500
```

Remaining balance

```
₹500
```

Since ₹500 is less than ₹1000,

Python executes

```python
raise InsufficientBalanceError(
    "Minimum balance of ₹1000 must be maintained."
)
```

The program stops and displays the custom exception.

---

# Handling the Exception

Instead of crashing the program,

we can handle it using `try-except`.

```python
class InsufficientBalanceError(Exception):
    pass


class BankAccount:

    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):

        if self.balance - amount < 1000:

            raise InsufficientBalanceError(
                "Minimum balance of ₹1000 must be maintained."
            )

        self.balance -= amount

        print(f"₹{amount} Withdrawn Successfully")


account = BankAccount(5000)

try:

    account.withdraw(4500)

except InsufficientBalanceError as error:

    print(error)
```

---

## Output

```text
Minimum balance of ₹1000 must be maintained.
```

# Real-Time Example 2: Student Age Validation

Suppose a university only allows students who are **18 years or older**.

---

## Custom Exception

```python
class InvalidAgeError(Exception):
    pass
```

---

## Student Class

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def register(self):

        if self.age < 18:

            raise InvalidAgeError(
                "Student must be at least 18 years old."
            )

        print("Registration Successful")
```

---

## Usage

```python
student = Student("Siva", 16)

student.register()
```

---

## Output

```text
InvalidAgeError:
Student must be at least 18 years old.
```

---

# Handling the Exception

```python
try:

    student = Student("Siva", 16)

    student.register()

except InvalidAgeError as error:

    print(error)
```

Output

```text
Student must be at least 18 years old.
```

# Flow of Execution

```
User

    │

    ▼

withdraw(4500)

    │

    ▼

Check Balance

    │

    ▼

Is Balance < ₹1000 ?

        │

   Yes ───────► raise Exception

        │

        ▼

except Block Executes

        │

        ▼

Display Error Message
```

# Difference Between `raise` and `Exception`

| `raise` | `Exception` |
|----------|-------------|
| Keyword | Base exception class |
| Used to throw an exception | Used to create custom exceptions |
| Stops normal execution | Parent of all exceptions |

Example

```python
raise ValueError("Invalid Input")
```

Here,

- `raise` throws the exception.
- `ValueError` is the exception class.

---

# Best Practices

- Create meaningful custom exception names.
- Inherit from the `Exception` class.
- Use custom exceptions only for business-specific rules.
- Always provide meaningful error messages.
- Handle exceptions using `try-except` whenever possible.

---

# Key Takeaways

- A **Custom Exception** is a user-defined exception.
- Custom exceptions are created by inheriting from the `Exception` class.
- The **`raise`** keyword is used to manually throw an exception.
- Use custom exceptions to enforce business rules.
- Handle custom exceptions using `try-except`.
- Custom exceptions make applications easier to debug, maintain, and understand.
