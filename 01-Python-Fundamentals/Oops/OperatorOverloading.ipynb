# Operator Overloading in Python

## Introduction

**Operator Overloading** is one of the most powerful features of Python's Object-Oriented Programming.

Normally, operators like:

- `+`
- `-`
- `*`
- `/`
- `==`
- `<`
- `>`
- `<=`
- `>=`

are designed to work with Python's built-in data types such as integers, floats, and strings.

However, by using **Magic Methods (Dunder Methods)**, we can define how these operators should behave for our own custom objects.

This feature is called **Operator Overloading**.

---

# Definition

**Operator Overloading** is the process of giving **special meaning** to an operator when it is used with objects of a custom class.

Python allows us to overload operators by implementing the corresponding **magic methods**.

For example,

- `+` → `__add__()`
- `-` → `__sub__()`
- `*` → `__mul__()`
- `==` → `__eq__()`

---

# Why Do We Need Operator Overloading?

Suppose Python already knows how to add integers.

```python
print(10 + 20)
```

Output

```text
30
```

Similarly,

```python
print("Hello " + " World")
```

Output

```text
Hello World
```

Python already knows how these operators work.

But suppose we create our own class.

```python
class Employee:

    def __init__(self, salary):
        self.salary = salary


emp1 = Employee(50000)
emp2 = Employee(40000)

print(emp1 + emp2)
```

Output

```text
TypeError:
unsupported operand type(s) for +: 'Employee' and 'Employee'
```

Python doesn't know how to add two Employee objects.

To solve this problem, we use **Operator Overloading**.

# Real-Time Example: Employee Salary Management System

Imagine you're developing an **Employee Management System**.

Each employee has a salary.

Sometimes HR wants to calculate the **combined salary** of two employees.

Instead of writing:

```python
total_salary = emp1.salary + emp2.salary
```

it would be much cleaner to write:

```python
total_salary = emp1 + emp2
```

This is possible using **Operator Overloading**.

---

# Without Operator Overloading

```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


emp1 = Employee("Siva", 50000)
emp2 = Employee("Teju", 40000)

print(emp1.salary + emp2.salary)
```

### Output

```text
90000
```

Works, but we are directly accessing object attributes.

---

# Trying to Add Objects

```python
print(emp1 + emp2)
```

### Output

```text
TypeError:
unsupported operand type(s) for +: 'Employee' and 'Employee'
```

Python doesn't know what **adding two employees** means.

---

# Using Operator Overloading

We overload the `+` operator using the `__add__()` magic method.

```python
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __add__(self, other):
        return self.salary + other.salary


emp1 = Employee("Siva", 50000)
emp2 = Employee("Teju", 40000)

print(emp1 + emp2)
```

### Output

```text
90000
```

---

# What Happens Internally?

When we write

```python
emp1 + emp2
```

Python automatically converts it into

```python
emp1.__add__(emp2)
```

Internally,

```
emp1 + emp2

      │

      ▼

emp1.__add__(emp2)

      │

      ▼

50000 + 40000

      │

      ▼

90000
```

You never call `__add__()` directly.

Python automatically calls it whenever the `+` operator is used.

---

# Complete Real-Time Example

```python
class Employee:

    def __init__(self, emp_id, name, salary):

        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def __add__(self, other):

        return self.salary + other.salary

    def __str__(self):

        return f"{self.name} - ₹{self.salary}"


emp1 = Employee(101, "Siva", 50000)
emp2 = Employee(102, "Teju", 40000)

print(emp1)
print(emp2)

print()

print("Total Salary :", emp1 + emp2)
```

---

# Output

```text
Siva - ₹50000
Teju - ₹40000

Total Salary : 90000
```

# Another Example: Shopping Cart

Suppose you have two shopping carts.

Cart 2

```text
5 Items
```

Instead of writing

```python
cart1.items + cart2.items
```

we can simply write

### Code

```python
class Cart:

    def __init__(self, items):
        self.items = items

    def __add__(self, other):
        return self.items + other.items


cart1 = Cart(3)
cart2 = Cart(5)

print(cart1 + cart2)
```

# Common Operator Overloading Methods

| Operator | Magic Method |
|-----------|--------------|
| `+` | `__add__()` |
| `-` | `__sub__()` |
| `*` | `__mul__()` |
| `/` | `__truediv__()` |
| `//` | `__floordiv__()` |
| `%` | `__mod__()` |
| `==` | `__eq__()` |
| `!=` | `__ne__()` |
| `<` | `__lt__()` |
| `>` | `__gt__()` |
| `<=` | `__le__()` |
| `>=` | `__ge__()` |

# Advantages of Operator Overloading

- Makes code more readable.
- Makes objects behave like built-in data types.
- Improves code maintainability.
- Reduces unnecessary method calls.
- Makes applications more intuitive.

---

# Best Practices

- Overload operators only when it makes logical sense.
- Keep operator behavior simple and predictable.
- Return meaningful values from overloaded methods.
- Avoid confusing or unexpected implementations.

# Another Real-Time Example: Bank Account

Suppose two bank accounts belong to the same customer.

Account 1

```
₹25,000
```

Account 2

```
₹40,000
```

Instead of writing

```python
account1.balance + account2.balance
```

we can simply write

```python
account1 + account2
```

---

# Code

```python
class BankAccount:

    def __init__(self, account_holder, balance):

        self.account_holder = account_holder
        self.balance = balance

    def __add__(self, other):

        return self.balance + other.balance


account1 = BankAccount("Siva", 25000)

account2 = BankAccount("Teju", 40000)

print("Total Balance :", account1 + account2)
```

---

# Output

```text
Total Balance : 65000
```

---

# What Happens Internally?

```
account1 + account2

        │

        ▼

account1.__add__(account2)

        │

        ▼

25000 + 40000

        │

        ▼

65000
```
