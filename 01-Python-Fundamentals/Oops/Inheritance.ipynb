# 1. Introduction to Inheritance

Inheritance is one of the four pillars of Object-Oriented Programming (OOP).

The four pillars are:

- Encapsulation
- Inheritance
- Polymorphism
- Abstraction

Inheritance allows one class to reuse the properties and methods of another class.

Instead of writing the same code again and again, we can inherit it.

---

## Definition

> Inheritance is the mechanism by which one class acquires the properties and methods of another class.

The existing class is called

- Parent Class
- Base Class
- Super Class

The new class is called

- Child Class
- Derived Class
- Sub Class

---

# 2. Why Do We Need Inheritance?

Imagine a company.

Every employee has

- Name
- Salary
- Employee ID

Instead of writing these variables in every class,

```
Developer

Manager

Tester

HR
```

We create one common class

```
Employee
```

Then every other class inherits it.

This avoids duplicate code.

This is called

**Code Reusability**

---

# Without Inheritance

```python
class Dog:
    def __init__(self):
        self.name = "Tom"

    def eat(self):
        print("Eating")


class Cat:
    def __init__(self):
        self.name = "Kitty"

    def eat(self):
        print("Eating")
```

Notice

The `eat()` method is duplicated.

---

# With Inheritance

```python
class Animal:

    def eat(self):
        print("Eating")


class Dog(Animal):
    pass


class Cat(Animal):
    pass
```

Now

Dog

Cat

Both automatically get `eat()`.

# 12. super()

`super()` gives access to parent class methods.

Example

```python
class Parent:

    def hello(self):
        print("Hello Parent")


class Child(Parent):

    def hello(self):

        super().hello()

        print("Hello Child")


c = Child()

c.hello()
```

Output

```
Hello Parent
Hello Child
```

# 14. Method Resolution Order (MRO)

Which method should execute first.

```python

    pass

    pass

    pass

d.show()
```

Output

```
A
```

---

View MRO

```python
print(D.mro())
```

Output

```
D
B
C
A
object
```

# 21. Best Practices

- Keep parent class generic.
- Child classes should extend behavior, not duplicate it.
- Use `super()` in constructors.
- Avoid deep inheritance chains.
- Prefer composition if inheritance is not a true "is-a" relationship.
- Override methods only when necessary.
- Use meaningful class names.
- Follow the Single Responsibility Principle (SRP).

# 22. Common Mistakes

### Forgetting `super().__init__()`

```python
class Parent:
    def __init__(self):
        self.name = "Python"


class Child(Parent):
    def __init__(self):
        pass


obj = Child()

print(obj.name)
```

Error

```
AttributeError
```

Fix

```python
class Child(Parent):
    def __init__(self):

        super().__init__()
```

---

### Calling Parent Method Incorrectly

Wrong

```python
Parent.show()
```

Correct

```python
super().show()
```

or

```python
self.show()
```

depending on the situation.

---

### Using Inheritance Unnecessarily

Not every relationship should use inheritance.

Example:

```
Car → Engine
```

A Car **has an** Engine, so composition is a better choice than inheritance.

# Summary

- Inheritance allows a class to reuse code from another class.
- The parent class contains common functionality, while child classes extend or customize it.
- Python supports five inheritance types:
  - Single
  - Multiple
  - Multi-Level
  - Hierarchical
  - Hybrid

1. Single Inheritance

A child class inherits from a single parent class.

````python
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def display(self):
        print(f"Engine Type: {self.enginetype}")


class Tesla(Car):
    def __init__(self, windows, doors, enginetype, is_selfdriving):
        super().__init__(windows, doors, enginetype)
        self.is_selfdriving = is_selfdriving

    def selfdriving(self):
        print(f"Self Driving: {self.is_selfdriving}")


tesla = Tesla(4, 4, "Electric", True)

tesla.display()
tesla.selfdriving()
````

2. Multiple Inheritance

A child class inherits from multiple parent classes.

````python
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def car_info(self):
        print(f"Engine Type: {self.enginetype}")


class AIFeatures:
    def self_driving(self):
        print("Supports Self Driving")

    def auto_parking(self):
        print("Supports Auto Parking")


class Tesla(Car, AIFeatures):
    def __init__(self, windows, doors, enginetype):
        Car.__init__(self, windows, doors, enginetype)


tesla = Tesla(4, 4, "Electric")

tesla.car_info()
tesla.self_driving()
tesla.auto_parking()
````

3. Multilevel Inheritance

Inheritance takes place over multiple levels.

````python
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def car_info(self):
        print(f"Engine Type: {self.enginetype}")


class ElectricCar(Car):
    def battery(self):
        print("Battery Capacity: 100 kWh")


class Tesla(ElectricCar):
    def self_driving(self):
        print("Tesla Supports Self Driving")


tesla = Tesla(4, 4, "Electric")

tesla.car_info()
tesla.battery()
tesla.self_driving()
````

4. Hierarchical Inheritance

Multiple child classes inherit from one parent class.

````python
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def display(self):
        print(f"Engine Type: {self.enginetype}")


class Tesla(Car):
    def self_driving(self):
        print("Tesla Self Driving")


class BMW(Car):
    def autopilot(self):
        print("BMW Driving Assistant")


tesla = Tesla(4, 4, "Electric")
bmw = BMW(4, 4, "Petrol")

tesla.display()
tesla.self_driving()

print()

bmw.display()
bmw.autopilot()
````

5. Hybrid Inheritance

Hybrid inheritance is a combination of two or more inheritance types (e.g., Multiple + Multilevel).



````python
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype

    def car_info(self):
        print(f"Engine Type: {self.enginetype}")


class ElectricCar(Car):
    def battery(self):
        print("100 kWh Battery")


class AIFeatures:
    def self_driving(self):
        print("Supports Self Driving")


class Tesla(ElectricCar, AIFeatures):
    pass


tesla = Tesla(4, 4, "Electric")

tesla.car_info()
tesla.battery()
tesla.self_driving()
````

| Inheritance Type | Description                                    | Example                                   |
| ---------------- | ---------------------------------------------- | ----------------------------------------- |
| Single           | One child inherits from one parent             | `Car → Tesla`                             |
| Multiple         | One child inherits from multiple parents       | `Car + AIFeatures → Tesla`                |
| Multilevel       | A child inherits from another child class      | `Car → ElectricCar → Tesla`               |
| Hierarchical     | Multiple child classes inherit from one parent | `Car → Tesla`, `Car → BMW`                |
| Hybrid           | Combination of multiple inheritance types      | `Car → ElectricCar`, `AIFeatures → Tesla` |
