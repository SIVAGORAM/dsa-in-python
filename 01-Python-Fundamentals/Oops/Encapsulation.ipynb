# Encapsulation and Abstraction in Python

## Introduction

**Encapsulation** and **Abstraction** are two of the four fundamental principles of **Object-Oriented Programming (OOP)**. They help developers design software that is **secure, modular, maintainable, and reusable**.

- **Encapsulation** focuses on **protecting data** by bundling data (attributes) and methods (functions) into a single unit (class) and controlling access to that data.
- **Abstraction** focuses on **hiding implementation details** and exposing only the essential functionality to the user.

Together, these principles reduce complexity, improve code organization, and make applications easier to develop and maintain.

# Encapsulation

## Definition

**Encapsulation** is the process of **bundling data (variables)** and the **methods (functions)** that operate on that data into a single unit called a **class**.

It also restricts direct access to an object's internal data, helping to prevent accidental modification and misuse.

In simple words,

> **Encapsulation is the process of hiding an object's internal data and allowing controlled access through methods.**

---

## Why Do We Need Encapsulation?

Encapsulation helps to:

- Protect object data from unauthorized access.
- Prevent accidental modification of data.
- Improve code security.
- Increase code maintainability.
- Promote modular programming.
- Improve code reusability.

---

## Real-World Example

Think about an **ATM Machine**.

When you withdraw money:

- You enter your PIN.
- The ATM verifies your account.
- It checks your balance.
- It dispenses cash.

You **cannot directly access or modify** the bank's database.

The internal implementation is hidden, and you interact only through the available options.

This is an example of **Encapsulation**.

---

## How Encapsulation Works

Encapsulation combines:

- Variables (Data)
- Methods (Functions)

inside a single class.

```
+----------------------+

```

Users interact with methods instead of modifying variables directly.

# Access Modifiers in Python

Python provides three types of access modifiers.

| Modifier | Syntax | Accessible |
|----------|--------|------------|
| Public | `name` | Anywhere |
| Protected | `_name` | Class and Subclass (by convention) |
| Private | `__name` | Inside the Class (name mangling) |

---

# Public Members

Public members can be accessed from anywhere.

```python
class Student:

    def __init__(self):
        self.name = "Siva"


student = Student()

print(student.name)
```

### Output

```text
Siva
```

---

# Protected Members

Protected members are prefixed with a single underscore (`_`).

By convention, they are intended for use within the class and its subclasses.

```python
class Student:

    def __init__(self):
        self._name = "Siva"


student = Student()

print(student._name)
```

### Output

```text
Siva
```

Although accessible, they **should not** be accessed directly outside the class.

---

# Private Members

Private members use double underscores (`__`).

Python performs **name mangling**, making them harder to access directly.

```python
class BankAccount:

    def __init__(self):
        self.__balance = 50000


account = BankAccount()

print(account.__balance)
```

### Output

```text
AttributeError
```

---

# Accessing Private Members

Private members can be accessed using methods.

```python
class BankAccount:

    def __init__(self):
        self.__balance = 50000

    def get_balance(self):
        return self.__balance


account = BankAccount()

print(account.get_balance())
```

### Output

```text
50000
```

---

# Name Mangling

Python changes private variable names internally.

Example

```python
class Bank:

    def __init__(self):
        self.__balance = 1000


bank = Bank()

print(bank._Bank__balance)
```

### Output

```text
1000
```

Although possible, this should **not** be used in normal programming.

---

# Getters and Setters

Instead of accessing variables directly, we use methods.

```python
class BankAccount:

    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance


account = BankAccount()

account.deposit(5000)

print(account.get_balance())
```

### Output

```text
5000
```

# Comparison of Access Modifiers

| Feature | Public | Protected | Private |
|----------|--------|-----------|----------|
| Syntax | `name` | `_name` | `__name` |
| Accessible Everywhere | ✅ Yes | ✅ Yes (by convention) | ❌ No |
| Accessible in Child Class | ✅ Yes | ✅ Yes | ❌ Directly No |
| Name Mangling | ❌ No | ❌ No | ✅ Yes |
| Appears in `dir()` | `name` | `_name` | `_ClassName__name` |
| Recommended Direct Access | ✅ Yes | ⚠️ Avoid | ❌ Never |

---

# Best Practices

✅ Use **Public** variables for non-sensitive data.

✅ Use **Protected** variables when subclasses need access.

✅ Use **Private** variables for sensitive data.

✅ Access private variables through **Getter** methods.

✅ Modify private variables through **Setter** methods.

✅ Inside the class, always use

```python
self.__variable
```

Never write

```python
self._ClassName__variable
```

because Python automatically performs **Name Mangling**.

---

# Key Takeaways

- Public variables are accessible from anywhere.
- Protected variables are accessible but intended for internal use.
- Private variables use **Name Mangling** to discourage direct access.
- Python changes `__variable` into `_ClassName__variable` internally.
- Use `dir()` to see how Python stores object attributes.
- Never access mangled names directly in your code.
- Always use Getter and Setter methods to access or modify private variables.
- Encapsulation improves security, maintainability, and code organization.

````python
"""
===========================================================
        ENCAPSULATION IN PYTHON - COMPLETE EXAMPLE
===========================================================

This example demonstrates:

1. Public Variables
2. Protected Variables
3. Private Variables
4. Getter Methods
5. Setter Methods
6. Name Mangling
7. Best Practices

-----------------------------------------------------------
"""

class Person:

    def __init__(self, name, age, gender):
        """
        Constructor

        Public Variable    : name
        Protected Variable : _gender
        Private Variable   : __age
        """

        # Public Variable
        self.name = name

        # Protected Variable (By Convention)
        self._gender = gender

        # Private Variable
        self.__age = age

    # =====================================================
    # Getter Methods
    # =====================================================

    def get_name(self):
        """
        Getter for Public Variable
        Although public variables can be accessed directly,
        we can still provide a getter.
        """
        return self.name

    def get_gender(self):
        """
        Getter for Protected Variable
        Recommended way to access protected variables.
        """
        return self._gender

    def get_age(self):
        """
        Getter for Private Variable

        Notice:
        We are using self.__age
        NOT self._Person__age

        Python automatically converts it internally.
        """
        return self.__age

    # =====================================================
    # Setter Methods
    # =====================================================

    def set_name(self, name):
        """
        Setter for Public Variable
        """
        self.name = name

    def set_gender(self, gender):
        """
        Setter for Protected Variable
        """
        self._gender = gender

    def set_age(self, age):
        """
        Setter for Private Variable

        Validation Example:
        Age should not be negative.
        """

        if age >= 0:
            self.__age = age
        else:
            print("Age cannot be negative!")

    # =====================================================
    # Display Method
    # =====================================================

    def display(self):
        """
        Display all details.
        """

        print("Name   :", self.name)
        print("Gender :", self._gender)
        print("Age    :", self.__age)


# ==========================================================
# Creating Object
# ==========================================================

person = Person("Siva", 22, "Male")


print("=" * 50)
print("ACCESSING PUBLIC VARIABLE")
print("=" * 50)

# Direct Access
print(person.name)

# Using Getter
print(person.get_name())


print("\n" + "=" * 50)
print("ACCESSING PROTECTED VARIABLE")
print("=" * 50)

# Direct Access (Possible but not recommended)
print(person._gender)

# Recommended Way
print(person.get_gender())


print("\n" + "=" * 50)
print("ACCESSING PRIVATE VARIABLE")
print("=" * 50)

# Direct Access
# Uncomment the below line to see the error

# print(person.__age)

# Output:
# AttributeError:
# 'Person' object has no attribute '__age'


# Recommended Way
print(person.get_age())


print("\n" + "=" * 50)
print("CHECKING HOW PYTHON STORES VARIABLES")
print("=" * 50)

print(dir(person))

"""
Notice the output.

Public Variable

name

Protected Variable

_gender

Private Variable

_Person__age

Python internally changes

__age

into

_Person__age

This process is called

Name Mangling
"""


print("\n" + "=" * 50)
print("ACCESSING PRIVATE VARIABLE USING NAME MANGLING")
print("=" * 50)

# Technically possible
print(person._Person__age)

"""
Although this works,

Never do this in real-world applications.

Always use Getter methods.
"""


print("\n" + "=" * 50)
print("UPDATING VALUES USING SETTER METHODS")
print("=" * 50)

person.set_name("Teju")
person.set_gender("Female")
person.set_age(24)

print(person.get_name())
print(person.get_gender())
print(person.get_age())


print("\n" + "=" * 50)
print("DISPLAYING UPDATED OBJECT")
print("=" * 50)

person.display()


print("\n" + "=" * 50)
print("VALIDATION USING SETTER")
print("=" * 50)

person.set_age(-10)

print(person.get_age())
````

Getter

A Getter is a method used to read (get) the value of a variable, especially a private variable.

Instead of accessing the variable directly, you access it through a getter method.

Syntax

````python
class Person:

    def __init__(self):
        self.__name = "Siva"

    def get_name(self):
        return self.__name
````

````python
person = Person()

print(person.get_name())
````

Why Use a Getter?

To provide controlled access to private data.

To hide the internal implementation of the class.

To protect sensitive information from direct access.

Setter

A Setter is a method used to update (set) the value of a variable, especially a private variable.

Instead of modifying the variable directly, you modify it through a setter method.

Syntax

````python
class Person:

    def __init__(self):
        self.__name = "Siva"

    def set_name(self, name):
        self.__name = name
````

````python
person = Person()

person.set_name("Teju")
````

Getter and Setter Together

````python
class Person:

    def __init__(self, name):
        self.__name = name

    # Getter
    def get_name(self):
        return self.__name

    # Setter
    def set_name(self, name):
        self.__name = name


person = Person("Siva")

# Getter
print(person.get_name())

# Setter
person.set_name("Teju")

# Getter
print(person.get_name())
````

Simple Analogy

Imagine a bank locker.

The locker contains your valuables (private data).
The key allows you to view or access the locker (Getter).
The bank manager updates the locker's records when needed (Setter).

You don't open the locker directly without authorization—you use the proper mechanism.

Difference Between Getter and Setter

| Getter                          | Setter                          |
| ------------------------------- | ------------------------------- |
| Used to read data               | Used to update data             |
| Returns the value of a variable | Changes the value of a variable |
| Usually starts with `get_`      | Usually starts with `set_`      |
| Example: `get_name()`           | Example: `set_name()`           |

n Simple Words
Getter = Get the value
Setter = Set (update) the value

They are commonly used with private variables to provide controlled and secure access to an object's data, which is a key part of encapsulation.
