# File Operations – Reading and Writing Files in Python

## Introduction

**File handling** is an essential part of programming that allows programs to store, retrieve, and manipulate data permanently. Unlike variables, which store data only while a program is running, files enable data to be saved and accessed even after the program has ended.

Python provides built-in functions and methods to perform various file operations, making it easy to work with both **text files** and **binary files**.

In this chapter, you will learn how to:

- Open files
- Read data from files
- Write data to files
- Append data to existing files
- Work with different file modes
- Handle both text and binary files
- Automatically close files using the `with` statement
- Follow best practices for file handling

## Why is File Handling Important?

File handling enables programs to:

- Store data permanently.
- Read previously saved information.
- Update existing files.
- Process large amounts of data efficiently.
- Exchange data between different applications.

## Types of Files

Python mainly works with two types of files:

### 1. Text Files

Text files store data in a human-readable format.

Examples:

- `.txt`
- `.csv`
- `.json`
- `.xml`
- `.html`

---

### 2. Binary Files

Binary files store data in binary (0s and 1s) format and are not human-readable.

Examples:

- `.jpg`
- `.png`
- `.pdf`
- `.mp3`
- `.mp4`
- `.exe`

## Key Takeaways

- File handling allows data to be stored permanently.
- Python provides built-in support for reading and writing files.
- Files are mainly classified as **text files** and **binary files**.
- The `with` statement is the recommended way to work with files because it automatically closes the file after use.
- File handling is widely used in real-world applications such as logging, data processing, configuration management, and report generation.

```python
## Read a whose file:-

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line

with open('example.txt','r') as file:
    for line in file:        
        print(line.strip()) ## strip() removes the newline character
```

```python
# Writing a file(overwriting)

with open('example.txt','w') as file:
    file.write('Hello world\n')
    file.write('this is a new line,')
    
```

```python
## Write a file without overwriting()

with open('example.txt','a') as file:
    file.write("Append operation taking place!\n")
```

```python
## writing a list of lines to a file

lines = ['first line \n','second line \n','third line \n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
### Binary Files

## Writing to a Binary file

data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
## Reading a Binary file


with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
## Read the content from source text file and write to a destination text file
## copying a text file

with open('example.txt','r') as source_file:
    content = source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
    
```

```python
## Read a text file and count the number of lines, words, and characters.
## Counting lines, words, and characters in a text file.

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readline()
        line_count = len(lines)
        word_count = sum(len(line.split())for lines in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines,words, characters = count_text_file(file_path)
print(f'Lines:{lines}, Words:{words}, Characters:{characters}')
```

The W+ mode in python is used to open a file for both reading and writing. if the file does not exist, it will be created. if the file exits, its content is truncated(i.e., the file is overwritten).

```python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello world\n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning

    file.seek(0)

    ## Read the content of the file
    content=file.read()
    print(content)
```
