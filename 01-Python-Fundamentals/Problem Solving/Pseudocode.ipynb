# Pseudocode

## Definition

A **Pseudocode** is an informal, human-readable way of describing the steps of an algorithm using plain English or a combination of natural language and programming-like statements.

It helps programmers focus on the logic of a solution without worrying about the syntax of any specific programming language.

---

## Characteristics of Pseudocode

- Written in simple and easy-to-understand language.
- Does **not** follow the syntax of any specific programming language.
- Describes the logical sequence of steps to solve a problem.
- Easy to read, write, and modify.
- Acts as a bridge between an algorithm and actual program code.

---

## Advantages of Pseudocode

- Easy for beginners to understand.
- Helps plan program logic before coding.
- Independent of any programming language.
- Makes debugging and problem-solving easier.
- Improves communication among developers.
- Can be easily converted into code.

---

## Disadvantages of Pseudocode

- No universal standard for writing pseudocode.
- Cannot be executed by a computer.
- May become lengthy for complex problems.
- Different people may write the same logic differently.

---

## General Guidelines

- Begin with **START** and end with **END**.
- Write one instruction per line.
- Use simple English statements.
- Indent statements inside decision-making and loops.
- Avoid language-specific syntax.

---

## Example 1: Add Two Numbers

```text
START

INPUT A, B

SUM ← A + B

PRINT SUM

END
```

---

## Example 2: Find the Maximum of Two Numbers

```text
START

INPUT A, B

IF A > B THEN
    PRINT A
ELSE
    PRINT B
END IF

END
```

---

## Pseudocode vs Flowchart

| Pseudocode | Flowchart |
|------------|-----------|
| Written in plain English | Graphical representation |
| Easy to write | Easy to visualize |
| Uses text-based steps | Uses symbols and arrows |
| Faster to modify | Takes more time to update |
| Focuses on logic | Focuses on process flow |

---

## When to Use Pseudocode

Use pseudocode when you want to:

- Plan a solution before coding.
- Explain an algorithm clearly.
- Design program logic.
- Communicate ideas with team members.
- Convert logic into any programming language.

# Pseudocode: Find the Sum of Two Numbers

## Problem

Write a pseudocode to input two numbers and display their sum.

### Pseudocode

```text
START

INPUT A, B

SUM ← A + B

PRINT "Sum =", SUM

END
```

---

## Explanation

1. Start the program.
2. Read two numbers `A` and `B`.
3. Calculate the sum using `SUM ← A + B`.
4. Display the value of `SUM`.
5. End the program.

---

## Example

### Input

```text
A = 10
B = 20
```

### Output

```text
Sum = 30
```

Alternative Version (Using Natural Language)

START

Read the first number (A)

Read the second number (B)

Add A and B

Store the result in SUM

Display SUM

END
