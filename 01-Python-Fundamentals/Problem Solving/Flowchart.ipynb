# Basics of Problem Solving

Problem solving is the process of analyzing a problem, understanding its requirements, and finding an efficient solution. Before writing code, it is important to plan the solution using tools such as **flowcharts** and **algorithms**.

---

# Flowchart

## Definition

A **Flowchart** is a graphical representation of a process, system, or algorithm. It uses standard symbols connected by arrows to illustrate the sequence of steps required to solve a problem or complete a task.

Flowcharts help visualize the logic of a program, making it easier to understand, design, debug, and communicate solutions before writing code.

---

## Characteristics of a Flowchart

- Uses standard symbols to represent different operations.
- Shows the sequence of steps using arrows.
- Starts with a **Start** symbol and ends with an **End** symbol.
- Represents the logical flow of a program.
- Easy to read and understand.
- Independent of any programming language.

---

## Advantages of Flowcharts

- Simplifies complex problems.
- Makes program logic easier to understand.
- Helps identify errors before coding.
- Improves communication among team members.
- Makes debugging and maintenance easier.
- Serves as good documentation for programs.

---

## Disadvantages of Flowcharts

- Can become large and difficult to manage for complex problems.
- Time-consuming to create.
- Updating a flowchart after changes can require significant effort.
- Not suitable for representing very large systems.

---

## Common Flowchart Symbols

| Symbol | Name | Purpose |
|---------|------|---------|
| ⭕ | Start/End (Terminator) | Represents the beginning or end of a process |
| ▭ | Process | Represents a calculation, instruction, or operation |
| ◇ | Decision | Represents a condition with Yes/No or True/False outcomes |
| ▱ | Input/Output | Represents input from the user or output to the screen |
| → | Flow Line | Indicates the direction of program execution |
| ○ | Connector | Connects different parts of a flowchart |

---

## Example

### Problem

Find the largest of two numbers.

### Steps

1. Start
2. Input two numbers (`A` and `B`)
3. Compare `A` and `B`
4. If `A > B`, print `A` as the largest number.
5. Otherwise, print `B` as the largest number.
6. End

---

## When to Use a Flowchart

Use a flowchart when you need to:

- Plan a program before coding.
- Understand complex logic.
- Explain an algorithm visually.
- Debug a process.
- Document the workflow of a program.

---

## Flowchart vs Algorithm

| Flowchart | Algorithm |
|------------|-----------|
| Graphical representation | Step-by-step written instructions |
| Uses symbols | Uses plain English or pseudocode |
| Easy to visualize | Easy to write and modify |
| Best for understanding logic | Best for implementation |

# Flowchart: Display the Maximum of Two Numbers

## Problem

Write a flowchart to find and display the maximum of two numbers.

### Algorithm

1. Start
2. Input two numbers `A` and `B`
3. Check if `A > B`
4. If **Yes**, display `A` as the maximum number.
5. Otherwise, display `B` as the maximum number.
6. End

---

## Flowchart

```text
            ┌──────────────┐
            │    START     │
            └──────┬───────┘
                   │
                   ▼
         ┌──────────────────┐
         │ Input A and B    │
         └────────┬─────────┘
                  │
                  ▼
          ┌────────────────┐
          │    A > B ?     │
          └──────┬─────────┘
             Yes │ No
                 │
        ┌────────▼──────┐      ┌───────────────┐
        │ Print A is    │      │ Print B is    │
        │ Maximum       │      │ Maximum       │
        └────────┬──────┘      └──────┬────────┘
                 │                    │
                 └──────────┬─────────┘
                            │
                            ▼
                   ┌────────────────┐
                   │      END       │
                   └────────────────┘
```

Example

Input

A = 15
B = 10

Output

Maximum Number = 15

Time Complexity:-

Time Complexity: O(1)
Space Complexity: O(1)

This is one of the first and most common flowchart problems for beginners and is frequently used to introduce decision-making using an if-else condition.

# Flowchart: Add Two Numbers

## Problem

Write a flowchart to input two numbers and display their sum.

### Algorithm

1. Start
2. Input two numbers `A` and `B`
3. Calculate `Sum = A + B`
4. Display `Sum`
5. End

---

## Flowchart

```text
                ┌──────────────┐
                │    START     │
                └──────┬───────┘
                       │
                       ▼
              ╱─────────────────╲
             ╱  Input A and B    ╲
             ╲                   ╱
              ╲─────────────────╱
                       │
                       ▼
              ┌─────────────────┐
              │  Sum = A + B    │
              └────────┬────────┘
                       │
                       ▼
              ╱─────────────────╲
             ╱   Print Sum       ╲
             ╲                   ╱
              ╲─────────────────╱
                       │
                       ▼
                ┌──────────────┐
                │     END      │
                └──────────────┘
```

---

## Example

### Input

```text
A = 10
B = 20
```

### Output

```text
Sum = 30
```

---

## Time Complexity

- **Time Complexity:** `O(1)`
- **Space Complexity:** `O(1)`
