# Introduction to Operators

Operators are special symbols or keywords used to perform operations on operands (values or variables).

Example

```python
x = 10
y = 20

print(x + y)
```

Here,

- `+` is the operator.
- `x` and `y` are operands.

---

# Why Operators are Important

Operators allow us to:

- Perform mathematical calculations.
- Compare values.
- Make decisions.
- Combine conditions.
- Manipulate data.
- Write efficient programs.

# Assignment Operators

Assignment operators assign values to variables.

| Operator | Example | Equivalent |
|----------|---------|-----------|
| = | x = 5 | Assign value |
| += | x += 2 | x = x + 2 |
| -= | x -= 2 | x = x - 2 |
| *= | x *= 2 | x = x * 2 |
| /= | x /= 2 | x = x / 2 |
| //= | x //= 2 | x = x // 2 |
| %= | x %= 2 | x = x % 2 |
| **= | x **= 2 | x = x ** 2 |

Example

```python
x = 10

x += 5

print(x)
```

Output

```
15
```

---

# Comparison Operators

Comparison operators compare two values.

They always return **True** or **False**.

| Operator | Meaning |
|----------|---------|
| == | Equal to |
| != | Not Equal to |
| > | Greater than |
| < | Less than |
| >= | Greater than or Equal to |
| <= | Less than or Equal to |

Example

```python
a = 10
b = 20

print(a == b)
print(a != b)
print(a < b)
print(a >= b)
```

Output

```
False
True
True
False
```

---

# Logical Operators

Logical operators combine multiple conditions.

| Operator | Meaning |
|----------|---------|
| and | Returns True if both conditions are True |
| or | Returns True if at least one condition is True |
| not | Reverses the Boolean value |

---

## AND

```python
age = 25
salary = 50000

print(age > 18 and salary > 30000)
```

Output

```
True
```

---

## OR

```python
print(age < 18 or salary > 30000)
```

Output

```
True
```

---

## NOT

```python
is_student = False

print(not is_student)
```

Output

```
True
```

# Identity Operators

Identity operators compare whether two variables refer to the same object.

| Operator | Meaning |
|----------|---------|
| is | Same object |
| is not | Different objects |

Example

```python
x = [1, 2]

y = x

print(x is y)
```

Output

```
True
```

---

# Membership Operators

Membership operators check whether a value exists inside a sequence.

| Operator | Meaning |
|----------|---------|
| in | Exists |
| not in | Does not exist |

Example

```python
course = "Python"

print("P" in course)

print("Java" not in course)
```

Output

```
True
True
```

# Operator Precedence

Python follows a precedence order.

Highest to Lowest

1. Parentheses `()`
2. Exponentiation `**`
3. Unary Operators (`+`, `-`, `not`)
4. Multiplication, Division, Floor Division, Modulus
5. Addition, Subtraction
6. Comparison Operators
7. Logical NOT
8. Logical AND
9. Logical OR

Example

```python
print(2 + 3 * 4)
```

Output

```
14
```

---

# Short-Circuit Evaluation

Python stops evaluating expressions when the result is already known.

Example

```python
age = 20

print(age > 18 or age / 0 > 1)
```

The second condition is never evaluated.

# Common Errors

## Division by Zero

```python
print(10 / 0)
```

Output

```
ZeroDivisionError
```

---

## Assignment instead of Comparison

Wrong

```python
if x = 10:
```

Correct

```python
if x == 10:
```

---

## Confusing `is` with `==`

Wrong

```python
a = 1000
b = 1000

print(a is b)
```

Correct

```python
print(a == b)
```

Use `==` to compare values and `is` to compare object identity.

---

# Best Practices

- Use parentheses to improve readability.
- Use `==` to compare values.
- Use `is` only for object identity.
- Avoid deeply nested logical expressions.
- Keep expressions simple and readable.
- Understand operator precedence before writing complex expressions.

````python
# Arithmetic Operators

a = 10
b = 5

# Addition
addition_result = a + b

# Subtraction
subtraction_result = a - b

# Multiplication
multiplication_result = a * b

# Division
division_result = a / b

# Floor Division
floor_division_result = a // b

# Modulus (Remainder)
modulus_result = a % b

# Exponentiation (Power)
exponentiation_result = a ** b

print("Addition:", addition_result)
print("Subtraction:", subtraction_result)
print("Multiplication:", multiplication_result)
print("Division:", division_result)
print("Floor Division:", floor_division_result)
print("Modulus:", modulus_result)
print("Exponentiation:", exponentiation_result)
````

````python
# Simple Calculator

print("===== Simple Calculator =====")

# Taking input from the user
num1 = float(input("Enter the first number: "))
operator = input("Enter an operator (+, -, *, /, //, %, **): ")
num2 = float(input("Enter the second number: "))

# Performing the operation
if operator == "+":
    print("Result:", num1 + num2)

elif operator == "-":
    print("Result:", num1 - num2)

elif operator == "*":
    print("Result:", num1 * num2)

elif operator == "/":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Error: Division by zero is not allowed.")

elif operator == "//":
    if num2 != 0:
        print("Result:", num1 // num2)
    else:
        print("Error: Division by zero is not allowed.")

elif operator == "%":
    if num2 != 0:
        print("Result:", num1 % num2)
    else:
        print("Error: Division by zero is not allowed.")

elif operator == "**":
    print("Result:", num1 ** num2)

else:
    print("Invalid operator! Please use +, -, *, /, //, %, or **.")
````
