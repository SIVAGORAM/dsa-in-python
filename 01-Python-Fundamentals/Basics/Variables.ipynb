# What is a Variable?

A **variable** is a named memory location used to store data.

In Python, variables are created automatically when a value is assigned to them. Unlike languages such as C or Java, Python does **not require explicit variable declaration**.

Think of a variable as a **container** that stores information which can be accessed and modified throughout a program.

---

# Why Do We Need Variables?

Variables help us to:

- Store data
- Reuse values
- Perform calculations
- Improve code readability
- Make programs dynamic

```python
## - Declaring and assigning Variables 
age =22
height = 5.9
name ="Siva"
is_student = True

## printing the variables
print("Age:", age)
print("Height:", height)
print("Name:", name)
print("Is Student:", is_student)
```

```python
## Naming Conventions
## Variable names should be descriptive and follow certain conventions:
# 1. Use lowercase letters and underscores for variable names (snake_case).
## 2. Avoid using reserved keywords as variable names.
## 3. Use meaningful names that convey the purpose of the variable.
## they must start with a letter or an '_' and contains letter, numbers and underscroes
## variables names are case sensitive

# valid vaiable names
first_name = "siva"
last_name = "goram"
```

```python
## Invalid variable names
# 1name = "siva"  # starts with a number
# class = "10th grade"  # uses a reserved keyword
# @age = 25  # contains an invalid character
```

```python
## Case Sensitivity
# Python is case-sensitive, meaning that variable names with different cases are treated as distinct variables.
name = "Siva"
Name = "Siva"
NAME = "Siva"
```

```python
## understanding Variable types
## Pythin is dynamically typed language, type of a variable is determined at runtime based on the value assigned to it.

age =22 #integer
height = 5.9 #float
is_student = True #boolean
name = "Siva" #string

print(type(age)) # <class 'int'>
print(type(height)) # <class 'float'>
print(type(is_student)) # <class 'bool'>
print(type(name)) # <class 'str'>
```

```python
## Type checking and conversion
# You can check the type of a variable using the type() function and convert between different types
type(height) # <class 'float'>
```

```python
age=22
print(type(age)) # <class 'int'>

# type conversion
age_str = str(age) # convert int to string
print(age_str) # "22"
print(type(age_str)) # <class 'str'>
```

```python
age ='25'
print(type(int(age))) # <class 'int'>
```

```python
name = "Siva"
int(name) # This will raise a ValueError since "Siva" cannot be converted to an integer
```

```python
height = 5.11
type(height) # <class 'float'>
```

```python
int(height) # This will raise a ValueError since 5.11 cannot be converted to an integer without losing information
```

```python
float(int(height)) # This will convert the float to an integer first and then back to a float, resulting in 5.0
```

```python
## Dynamic Typing
# Python is dynamically typed, meaning that you can reassign variables to different types without explicitly declaring
## python allows the type of a variables to changes as the program executes
var=10
print(var,type(var)) # 10 <class 'int'>

var="Hello"
print(var,type(var)) # Hello <class 'str'>

var=3.14
print(var,type(var)) # 3.14 <class 'float'>

```

```python
##input and output
# Python provides the input() function to read user input from the console. The input is always
# returned as a string, so you may need to convert it to the appropriate type if needed.

age=input("Enter your age: ")
print("Your age is:", type(age),age)
```

```python
age=int(input("Enter your age: "))
print("Your age is:", type(age),age)
```

```python
## Simple calculator that takes two numbers as input and performs addition, subtraction, multiplication, and division.
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)
print("Integer Division:", num1 // num2)
```
