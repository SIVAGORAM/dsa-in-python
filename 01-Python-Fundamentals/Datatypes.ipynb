# What are Data Types?

A **data type** is a classification that specifies what kind of value a variable can store.

Data types tell the Python interpreter how data should be stored, processed, and manipulated.

Every value in Python belongs to a specific data type.

Example:

```python
age = 22
name = "Siva"
salary = 35000.50
is_student = True
```

Here,

- `22` is an Integer
- `"Siva"` is a String
- `35000.50` is a Float
- `True` is a Boolean

---

# Importance of Data Types

Data types play an important role in programming because they:

- Store data efficiently.
- Determine the operations that can be performed.
- Improve program performance.
- Reduce programming errors.
- Make code easier to understand and maintain.

Example:

```python
age = 22
```

Since `age` is an integer, mathematical operations can be performed on it.

```python
print(age + 5)
```

Output

```
27
```

# Categories of Data Types

Python provides several built-in data types.

They can be broadly classified into:

## Basic Data Types

- Integer (`int`)
- Float (`float`)
- String (`str`)
- Boolean (`bool`)

## Advanced Data Types

- List (`list`)
- Tuple (`tuple`)
- Set (`set`)

---

# Basic Data Types

## 1. Integer (int)

Integers are whole numbers without decimal points.

Examples:

```python
age = 22
marks = 95
temperature = -5

print(type(age))
```

Output

```
<class 'int'>
```

---

## 2. Float (float)

Floats represent numbers with decimal points.

Examples:

```python
price = 199.99
height = 5.8
pi = 3.14159

print(type(price))
```

Output

```
<class 'float'>
```

---

## 3. String (str)

Strings are sequences of characters enclosed in single, double, or triple quotes.

Examples:

```python
name = "Siva"

city = 'Kurnool'

message = """Welcome to Python"""
```

Output

```python
print(name)
```

```
Siva
```

---

## 4. Boolean (bool)

Booleans represent logical values.

There are only two Boolean values:

- True
- False

Example

```python
is_logged_in = True

is_admin = False

print(type(is_logged_in))
```

Output

```
<class 'bool'>
```

---

# Advanced Data Types

## 1. List

A list is an ordered, mutable collection of items.

Lists can store different types of data.

Example

```python
numbers = [10, 20, 30, 40]

student = ["Siva", 22, True]

print(numbers)
```

Output

```
[10, 20, 30, 40]
```

Lists are mutable, meaning their elements can be modified.

---

## 2. Tuple

A tuple is an ordered, immutable collection.

Example

```python
coordinates = (10, 20)

days = ("Mon", "Tue", "Wed")

print(coordinates)
```

Output

```
(10, 20)
```

Tuples cannot be modified after creation.

---

## 3. Set

A set is an unordered collection of unique elements.

Duplicate values are automatically removed.

Example

```python
numbers = {1, 2, 3, 3, 4, 4, 5}

print(numbers)
```

Output

```
{1, 2, 3, 4, 5}
```

Sets are useful when unique values are required.

---

## 4. Dictionary

A dictionary stores data as key-value pairs.

Example

```python
student = {
    "name": "Siva",
    "age": 22,
    "city": "Kurnool"
}

print(student)
```

Output

```
{'name': 'Siva', 'age': 22, 'city': 'Kurnool'}
```

Values are accessed using their keys.

```python
print(student["name"])
```

Output

```
Siva
```

---

# None Type

Python provides a special data type called **NoneType**.

It represents the absence of a value.

Example

```python
result = None

print(result)

print(type(result))
```

Output

```
None

<class 'NoneType'>
```

# Mutable vs Immutable Data Types

| Mutable | Immutable |
|----------|-----------|
| List | Integer |
| Dictionary | Float |
| Set | Boolean |
| | String |
| | Tuple |
| | None |

**Mutable:** Can be modified after creation.

**Immutable:** Cannot be modified after creation.

# Type Checking

Use the `type()` function to determine the data type of a variable.

Example

```python
age = 22

print(type(age))

name = "Siva"

print(type(name))
```

Output

```
<class 'int'>

<class 'str'>
```

---

# Type Conversion

Python allows converting one data type into another.

Example

```python
age = "22"

age = int(age)

print(age)

print(type(age))
```

Output

```
22

<class 'int'>
```

Common Conversion Functions

```python
int()

float()

str()

bool()

list()

tuple()

set()

dict()
```

Example

```python
number = 10

print(float(number))

print(str(number))
```

Output

```
10.0

10
```

# Common Errors

## Adding a String and an Integer

```python
age = "22"

print(age + 5)
```

Output

```
TypeError
```

Correct

```python
print(int(age) + 5)
```

---

## Accessing a Dictionary with an Invalid Key

```python
student = {"name": "Siva"}

print(student["age"])
```

Output

```
KeyError
```

---

## Modifying a Tuple

```python
numbers = (1, 2, 3)

numbers[0] = 10
```

Output

```
TypeError
```

- Convert data types only when necessary.
- Avoid unnecessary conversions.
- Use lists for mutable collections.
- Use tuples for fixed data.
- Use sets for unique values.
- Use dictionaries to represent related data.

# Key Takeaways

- Data types define the kind of value a variable can store.
- Python provides several built-in data types.
- Basic data types include `int`, `float`, `str`, and `bool`.
- Advanced data types include `list`, `tuple`, `set`, and `dict`.
- Use `type()` to check a variable's data type.
- Use built-in functions such as `int()`, `float()`, and `str()` for type conversion.
- Lists, dictionaries, and sets are mutable.
- Integers, floats, strings, tuples, booleans, and `None` are immutable.
- Choosing the correct data type makes programs more efficient, readable, and maintainable.
