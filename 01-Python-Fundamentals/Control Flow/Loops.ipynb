# 1. Introduction to Loops

A loop is a programming construct that repeatedly executes a block of code until a specified condition is met.

Instead of writing the same statement multiple times, loops allow us to automate repetitive tasks.

Without loops:

```python
print("Hello")
print("Hello")
print("Hello")
print("Hello")
print("Hello")
```

With a loop:

```python
for i in range(5):
    print("Hello")
```

# 4. The for Loop

The `for` loop is used to iterate over a sequence.

Syntax

```python
for variable in iterable:
    statements
```

Example

```python
for i in range(5):
    print(i)
```

Output

```
0
1
2
3
4
```

---

# 5. Iterating over range()

The `range()` function generates a sequence of numbers.

## range(stop)

```python
for i in range(5):
    print(i)
```

---

## range(start, stop)

```python
for i in range(1, 6):
    print(i)
```

---

## range(start, stop, step)

```python
for i in range(0, 11, 2):
    print(i)
```

Output

```
0
2
4
6
8
10
```

# 13. The while Loop

A `while` loop executes as long as its condition is True.

Syntax

```python
while condition:
    statements
```

Example

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

# 15. Loop Control Statements

## break

Stops the loop immediately.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

---

## continue

Skips the current iteration.

```python
for i in range(6):
    if i == 3:
        continue
    print(i)
```

---

## pass

Acts as a placeholder.

```python
for i in range(5):
    if i == 2:
        pass
    print(i)
```

---

# 16. Nested Loops

A loop inside another loop.

```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

Output

```
0 0
0 1
0 2
1 0
1 1
1 2
2 0
2 1
2 2
```

# 17. The else Clause with Loops

Python allows an `else` block with loops.

```python
for i in range(3):
    print(i)
else:
    print("Loop Finished")
```

If the loop exits using `break`, the `else` block is skipped.

# 19. Common Errors

## Forgetting to Update the Counter

```python
count = 1

while count <= 5:
    print(count)
```

Results in an infinite loop.

---

## Incorrect Indentation

```python
for i in range(5):
print(i)
```

Raises an `IndentationError`.

---

## Modifying a List While Iterating

```python
numbers = [1, 2, 3]

for num in numbers:
    numbers.remove(num)
```

Avoid modifying collections while iterating over them.

# 20. Best Practices

- Use meaningful variable names.
- Use `for` loops when the number of iterations is known.
- Use `while` loops when the stopping condition is unknown.
- Avoid infinite loops.
- Keep nesting levels low.
- Use `break` and `continue` carefully.

While Loop

The While Loop Continues to execute as long as the condition is true.

````python
## While Loop
count = 0

while count<5:
    print(count)
    count=count+1
    
````

Loop Control Statements

- Break
- the break statement exits the loop permanentaly



````python
## break statement

for i in range(10):
    if i == 5:
        break
    print(i)
````

Continue

- The continue statement skips the current iteration and continues with the next.



````python
## Continue

for i in range(10):
    if i%2==0:
        continue
    print(i)
````

Pass
- The pass statement is a null operation, it does nothing.

````python
## Pass

for i in range(5):
    if i ==3:
        pass
    print(i)
        
````

Nested Loops
- A Loop inside a loop


````python
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
````

````python
## Example:- Calculate the sum of first N natural numbers using a while and for loop

## while Loop

n = 10
sum = 0
count = 1

while count<=n:
    sum = sum+count
    count=count+1
print("sum of first 10 natural number:",sum)
````

````python
n=10
sum=0
for i in range(11):
    sum=sum+i
print(sum)
````

````python
## Example:- Prime Numbers between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
````
