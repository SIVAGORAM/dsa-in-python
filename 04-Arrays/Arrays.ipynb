# Arrays

## Definition

An **array** is a linear data structure that stores a collection of elements in **contiguous memory locations**.

Each element is stored at a unique **index**, allowing efficient access to any element using its index.

In most programming languages such as **C, C++, and Java**, all elements in an array must be of the **same data type** (e.g., `int`, `float`, `char`).

> **Note:** Python does not have a built-in array data structure like C or Java. Instead, Python commonly uses **lists**, which can store elements of different data types. Python also provides an `array` module for creating arrays that store only elements of the same type.

---

## Characteristics of Arrays

- Stores elements in **contiguous memory locations**.
- Elements are accessed using an **index**.
- Indexing starts from **0** in most programming languages.
- Supports **fast random access** using indices.
- In traditional arrays, all elements must have the **same data type**.
- Array size is usually **fixed** after creation (except for dynamic arrays like Python lists or C++ vectors).

---

## Advantages of Arrays

- Fast access to elements using indexing (`O(1)`).
- Easy to traverse using loops.
- Efficient memory usage.
- Suitable for storing multiple values of the same type.

---

## Disadvantages of Arrays

- Fixed size in many programming languages.
- Insertion and deletion can be expensive (`O(n)`).
- Memory may be wasted if the array size is larger than required.
- Traditional arrays can store only one data type.

## Arrays in Python

Python does not use traditional arrays as its primary data structure. Instead, it commonly uses **lists**, which are more flexible and versatile.

Unlike traditional arrays, Python lists:

- Can store elements of different data types.
- Can grow or shrink dynamically.
- Provide many built-in methods for easy data manipulation.

However, Python also provides support for arrays through modules such as:

- **`array`** – Stores elements of the same data type and is more memory-efficient than lists for homogeneous data.
- **NumPy (`numpy`)** – Provides powerful, high-performance multidimensional arrays and mathematical operations, making it the standard library for scientific computing and data analysis.

### Example: Python List

```python
numbers = [10, 20, 30, 40, 50]
```

### Example: `array` Module

```python
from array import array

numbers = array('i', [10, 20, 30, 40, 50])
```

### Example: NumPy Array

```python
import numpy as np

numbers = np.array([10, 20, 30, 40, 50])
```

> **Note:** For most Python programming and DSA problems, **lists** are used instead of traditional arrays because they are dynamic, flexible, and easier to work with.

# Difference Between Array and List in Python

| Feature | Array | List |
|---------|-------|------|
| **Data Type** | Homogeneous (stores elements of the same data type) | Heterogeneous (can store elements of different data types) |
| **Size** | Usually fixed (traditional arrays) | Dynamic (can grow and shrink as needed) |
| **Memory Usage** | More memory-efficient for homogeneous data | Uses more memory due to flexibility |
| **Performance** | Faster for numerical operations and large homogeneous datasets | Slightly slower for numerical computations |
| **Flexibility** | Less flexible | Highly flexible |
| **Built-in Methods** | Fewer methods | Rich set of built-in methods |
| **Usage in Python** | Available through the `array` module or NumPy | Built-in data structure (`list`) |
| **Supports Mixed Data Types** | ❌ No | ✅ Yes |
| **Primary Use Case** | Numerical data and scientific computing | General-purpose programming and DSA |
| **Example** | `array('i', [1, 2, 3])` | `[1, 2, 3]` |

## Which One Should You Use?

### Use a **List** when:

- You are solving DSA problems.
- You need to store different data types.
- You need a dynamic data structure.
- You want to use Python's built-in features.

### Use an **Array** when:

- All elements are of the same data type.
- Memory efficiency is important.
- You are working with large numerical datasets.
- You are using scientific libraries like NumPy.

---

## Key Takeaways

- **Lists** are Python's default and most commonly used sequence type.
- **Arrays** are more memory-efficient for homogeneous numerical data.
- For **DSA and coding interviews**, you will almost always use **lists**.
- For **data science, machine learning, and numerical computing**, **NumPy arrays** are the preferred choice.
